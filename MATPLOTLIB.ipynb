```python
"""

Write a code to plot the Scores of 5 students as Pie Chart as given following.

Where:--
    
    Write the name of students corresponding to theire score.
    Shadow of pie chart should be True.
    The highest score should be exploded by 0.1.
    Start anngle should be start from 0 Degree.
    The percentage should be go to two digit after point.
    The size of the graph should be (10,10)

Inputs:--

    title = 'STUDENTS SCORE'
    
    student1 : 475
    student2 : 423
    student3 : 265
    student4 : 319
    student5 : 485

"""

import matplotlib.pyplot as plt


# Pie chart, where the slices will be ordered and plotted counter-clockwise:
names = ['Student1', 'Student2', 'Student3', 'Student4', 'Student5']

scores = [475, 423, 265, 319, 485 ]
explode = (0, 0, 0, 0, 0.1)  # only "explode" the 2nd slice (i.e. 'Hogs')
plt.figure(figsize = (10,10))
plt.pie(scores, explode = explode, labels = names, autopct='%1.2f%%', shadow = True, startangle = 0)
plt.title('STUDENTS SCORE')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
"""

Write a code to add the name of student out of graph like legend of barchat graph.
        
    Scores should be also presents in corresponding slice of pie chart.
    Size of graph should be of (10, 10)
    Fontsize should be 15

Inputs:--

    data = ['Student1 475', 'Student2 423', 'Student3, 265', 'Student4 319', 'Student5 485']

"""

import numpy as np
import matplotlib.pyplot as plt


data = ['Student1 475', 'Student2 423', 'Student3, 265', 'Student4 319', 'Student5 485']

# Spliting name and score from data.
names = [x.split()[0] for x in data]
scores = [float(x.split()[1]) for x in data]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} score)".format(pct, absolute)

# Figure size
plt.subplots(figsize=(10, 10))

# Highlighting the highest slice of piechart.
explode = (0, 0, 0, 0, 0.1)


# Plot the pie chart.
wedges, texts, autotexts = plt.pie(scores, autopct = lambda pct: func(pct, scores), textprops = dict(color="w"))

# plot the legend
# leg = plt.legend(wedges, names, title = 'Name of studnets', loc = "center left", bbox_to_anchor = (1, 0, 0.5, 1), fontsize = 15)

plt.setp(autotexts, size = 15, weight = "bold")

plt.title('Scores of Students', fontsize = 15, pad = -30)

leg = plt.legend(fontsize = 'large', labels = names)
leg.set_title("title", prop = {'size':'x-large'})

plt.show()
```

```python
"""

Write a code to plot the donat piechart.


"""

import numpy as np
import matplotlib.pyplot as plt


data = ['Student1 475', 'Student2 423', 'Student3, 265', 'Student4 319', 'Student5 485']


# Spliting name and score from data.
names = [x.split()[0] for x in data]
scores = [float(x.split()[1]) for x in data]


fig, ax = plt.subplots(figsize=(8, 8), subplot_kw = dict(aspect="equal"))


wedges, texts = ax.pie(scores, wedgeprops = dict(width = 0.5), startangle = 0)

bbox_props = dict(boxstyle = "square, pad=0.3", fc = "w", ec = "k", lw = 0.72)
kw = dict(arrowprops=dict(arrowstyle = "-"), bbox=bbox_props, zorder=0, va="center")


for i, p in enumerate(wedges):

    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(names[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)


ax.set_title("Students Record")
plt.show()
```

```python
"""

Write a code to plot a nested donut pie chart.


"""


import numpy as np
import matplotlib.pyplot as plt

vals = np.array([[95, 95, 95, 95, 95], [90, 85, 80, 80,75], [89, 77, 80, 90, 67], [67, 90, 67, 45, 89],[90, 78, 56, 75, 95]])

fig, ax = plt.subplots(figsize = (10,10))

size = 0.3
# vals = np.array([[60, 32, 40, 79], [37, 40, 56, 56], [98, 56, 78, 87]])

# cm = plt.get_cmap("tab20c")

# Create colors
a, b, c, d, e = [plt.cm.Blues, plt.cm.Reds, plt.cm.Greens, plt.cm.Purples, plt.cm.Oranges]

outer_colors = [a(0.7), b(0.7), c(0.7), d(0.7), e(0.7)]
inner_colors = [a(0.6), a(0.5), a(0.4), a(0.3), a(0.2), 
                b(0.6), b(0.5), b(0.4), b(0.3), b(0.2), 
                c(0.6), c(0.5), c(0.4), c(0.3), c(0.2),
                d(0.6), d(0.5), d(0.4), d(0.3), d(0.2),
                e(0.6), e(0.5), e(0.4), e(0.3), e(0.2)]

ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
     wedgeprops=dict(width=size, edgecolor='k'))

ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='k'))



ax.set(aspect="equal", title='Pie plot with `ax.pie`')
plt.show()
```

```python
"""

Write a code to scatter data in space(3d visualiztion).

"""

import matplotlib.pyplot as plt
import numpy as np

x = random.sample(range(0, 50), 50)
y = random.sample(range(0, 50), 50)
z = random.sample(range(0, 50), 50)

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, marker = '.', s = 500, c = 'red')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
